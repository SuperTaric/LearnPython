```python
from zipfile import ZipFile

with open('a.txt', 'w') as myfile:
    myfile.write('Hello ZipFile')

with ZipFile('a.zip', 'w') as myzip:
    myzip.write('a.txt')

with ZipFile('a.zip', 'r') as myzip:
    with myzip.open('a.txt') as myfile:
        print(myfile.read())
```

```python
from sympy import Symbol, Derivative
x = Symbol('x')
y = x * x + 3 * x + 2
d = Derivative(y, x)
d.doit()
```

```python
d.doit().subs({x: 10})
```

```python
from sympy.plotting import plot
plot(y, (x, -30, 30))
```

```python
from sympy import Symbol, limit, oo

n = Symbol('n')
y = (1 + 1/n) ** n

y_limit = limit(y, n, oo)
print(y_limit)
```
